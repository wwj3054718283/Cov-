```python
import time
def get_tencent_data():
    url = "https://view.inews.qq.com/g2/getOnsInfo?name=disease_h5"
    headers = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36"
    }
    r = requests.get(url,headers)
    res = json.loads(r.text)
    data_all = json.loads(res['data'])
    
    history = {} #历史数据
    ds = data_all["lastUpdateTime"]
#     tup = time.strptime(ds,"%Y.%m.%d")
#     ds = time.strftime("%Y-%m-%d",tup)
    confirm = data_all["chinaTotal"]["confirm"]
    heal = data_all["chinaTotal"]["heal"]
    dead = data_all["chinaTotal"]["dead"]
    nowConfirm = data_all["chinaTotal"]["nowConfirm"]
    suspect = data_all["chinaTotal"]["suspect"]
    history[ds] = {"confirm": confirm,"nowConfirm":nowConfirm, "suspect": suspect, "heal": heal, "dead": dead}
    
    details = [] #当日详情数据
    updata_time = data_all["lastUpdateTime"]
    data_province = data_all["areaTree"][0]["children"]
    for pro_infos in data_province:
        province = pro_infos["name"]
        for city_infos in pro_infos["children"]:
            city = city_infos["name"]
            confirm = city_infos["total"]["confirm"]
            nowConfirm = city_infos["total"]["nowConfirm"]
            heal = city_infos["total"]["heal"]
            dead = city_infos["total"]["dead"]
            details.append([updata_time,province,city,confirm,nowConfirm,heal,dead])
    
    return history,details
```

```python
def get_conn():
    conn = pymysql.connect(host="127.0.0.1",user="root",password="wwj520hy",db="cov",charset="utf8")
    cursor = conn.cursor()
    return conn, cursor

def close_conn(conn, cursor):
    if cursor:
        cursor.close()
    if conn:
        conn.close()
```

```python
def update_details():
    cursor = None
    conn = None
    try:
        li = get_tencent_data()[1]
        conn, cursor = get_conn()
        sql = "insert into details(update_time,province,city,confirm,nowConfirm,heal,dead) values(%s,%s,%s,%s,%s,%s,%s)"
        sql_query = 'select %s=(select update_time from details order by id desc limit 1)'#对比当前最大时间戳
        cursor.execute(sql_query,li[0][0])
        if not cursor.fetchone()[0]:
            print(f"{time.asctime()}开始更新最新数据")
            for item in li:
                cursor.execute(sql, item)
            conn.commit()
            print(f"{time.asctime()}更新最新数据完成")
        else:
            print(f"{time.asctime()}已是最新数据！")
    except:
        traceback.print_exc()
    finally:
        close_conn(conn, cursor)
```

```python
def insert_history():
    cursor = None
    conn = None
    try:
        dic = get_tencent_data()[0]
        print(f"{time.asctime()}开始插入历史数据")
        conn, cursor = get_conn()
        sql = "insert into history values(%s,%s,%s,%s,%s,%s)"
        for k, v in dic.items():
            cursor.execute(sql,[k, v.get("confirm"),v.get("nowConfirm"),v.get("suspect"),v.get("heal"),v.get("dead")])
        conn.commit()
        print(f"{time.asctime()}插入历史数据完成")
    except:
        traceback.print_exc()
    finally:
        close_conn(conn, cursor)
```

## 爬取百度热搜数据

```python
from selenium.webdriver import Chrome,ChromeOptions
from selenium.webdriver.common.by import By
```

```python
def get_baidu_hot():
    
    option = ChromeOptions()
    option.add_argument("--headless")#不启动浏览器
    option.add_argument("--no-sandbox")#Linux系统禁用参数

    url="https://top.baidu.com/board?tab=realtime"
    browser = Chrome(options = option)
    browser.get(url)
    # print(browser.page_source)
    but = browser.find_elements
    c = browser.find_elements(By.XPATH,'//*[@id="sanRoot"]/main/div[2]/div/div[2]/div/div[2]/a/div[1]')
    context = [i.text for i in c]
    return context
```

```python
get_baidu_hot()
```

```python
def updata_hotsearch():
    cursor = None
    conn = None
    try:
        context = get_baidu_hot()
        print(f"{time.asctime()}开始更新热搜数据")
        conn, cursor = get_conn()
        sql = "insert into hotsearch(dt,content) values(%s,%s)"
        ts = time.strftime("%Y-%m-%d %X")
        for i in context:
            cursor.execute(sql,(ts, i))
        conn.commit()
        print(f"{time.asctime()}数据更新完成")
    except:
        traceback.print_exc()
    finally:
        close_conn(conn, cursor)
```
